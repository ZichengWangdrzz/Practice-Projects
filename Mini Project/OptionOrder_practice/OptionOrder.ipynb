### 3. Changing data type of "OrderEventTypeOmIn" into int for operation

```python
df.dtypes
# '0' stands for 'CANCEL'
# '1' stands for 'FILL'
df = df.replace('CANCEL','0')
df = df.replace('FILL','1')
df = df[pd.to_numeric(df['OrderEventTypeOmIn'], errors = 'coerce').notnull()]
df['OrderEventTypeOmIn'] = df['OrderEventTypeOmIn'].astype('int')
df.dtypes
df
```

### 4.Generate independent and dependent variables

```python
df.columns

x_df = df[['Strike','Size', 'Price','MarketSize']]

# Indepedent variable
x = np.asarray(x_df)

# Dependent variable
y = np.asarray(df["OrderEventTypeOmIn"])
```

### 5. Divide the data as Training (80%)/Test dataset (20%)

```python
'''
df --> Train (80%) / Test (20%)
Train (x,y) ##x is a 2D array consisting of all independent variables #y is an 1D array
Test (x,y)
'''

#import sklearn, 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 1) #seed value
x_train.shape
x_test.shape

##scaling = MinMaxScaler(feature_range=(-1,1)).fit(x_train)
##x_train = scaling.transform(x_train)
##x_test = scaling.transform(x_test)
```

### 7. Improve SVM
After taking exactly 426m 4.5s of training my model. I decided to search online for help. 

The following improvment on SVM is cited as following:

Alexander, B. (2015, Aug 15). Making SVM run faster in python. stackoverflow. https://stackoverflow.com/questions/31681373/making-svm-run-faster-in-python




```python
import time
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn import datasets
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

# Utilitize the fastest SVC Classifier
start = time.time()
clf = OneVsRestClassifier(SVC(kernel='linear', probability=True))
clf.fit(x_train, y_train)
end = time.time()
print ("Single SVC", end - start, clf.score(x_train,y_train))
proba = clf.predict_proba(x_train)

n_estimators = 10
start = time.time()
clf = OneVsRestClassifier(BaggingClassifier(SVC(kernel='linear', probability=True), max_samples=1.0 / n_estimators, n_estimators=n_estimators))
clf.fit(x_train, y_train)
end = time.time()
print ("Bagging SVC", end - start, clf.score(x_train,y_train))
proba = clf.predict_proba(x_train)

start = time.time()
clf = RandomForestClassifier(min_samples_leaf=20)
clf.fit(x_train, y_train)
end = time.time()
print ("Random Forest", end - start, clf.score(x_train,y_train))
proba = clf.predict_proba(x_train)
```

It is noticiable that the "RandomForestClassifier" has not only the fastest speed, also the most accurate results

```python
y_predicted = clf.predict(x_test)
                
```

### 8. Evaluation of the model
Based on the classification report. I believe that I have just trained a nearly perfect SVM model, which is trustworthy. This set the base of processing a reliable coefficient analysis.

```python
from sklearn.metrics import classification_report

# Compare true and predicted value of y 
classification_report(y_test, y_predicted, output_dict= True)
```
